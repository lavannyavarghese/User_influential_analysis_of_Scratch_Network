```python
df=pd.read_csv("user.csv")



```

```python
df.head()
```

```python
G = nx.from_pandas_edgelist(df, 
                            source='user',
                            target='follower',
                            create_using=nx.DiGraph())
```

```python
nx.info(G)
```

```python
nx.draw_networkx(G)
```

```python
import operator
pr = nx.pagerank(G)
sorted_x = sorted(pr.items(), key=operator.itemgetter(1),reverse=True)
print('user_ranking')
print(sorted_x[:20])
```

```python
from operator import itemgetter
from IPython.display import HTML
from IPython.core.display import display
```

```python
print ("Degree Centrality")
print (sorted(nx.degree_centrality(G).items(),key=itemgetter(1), reverse=True)[:20])
```

```python
print ("Betweenness Centrality")
print (sorted(nx.betweenness_centrality(G).items(), key=itemgetter(1), reverse=True)[:20])
```

```python
print ("Closeness Centrality")
print (sorted(nx.closeness_centrality(G).items(), key=itemgetter(1), reverse=True)[:20])
```

```python
hub, authority = nx.hits(G)
sorted_hub = sorted(hub.items(), key=itemgetter(1),reverse=True)
print("Most Influential Users based on Hub Score")
print(sorted_hub[:20])
print()
sorted_authority = sorted(authority.items(), key=itemgetter(1),reverse=True)
print("Most Influential Users based on Authority Score")
print(sorted_authority[:20])
```
